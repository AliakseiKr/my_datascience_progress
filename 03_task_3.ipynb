```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\lesha\Downloads\train.csv")

plt.rcParams["font.family"] = "Times New Roman"
palette = {0: "lightskyblue", 1: "lightpink"}

# Graph 1  Survival by gender
plt.figure()
graph1 = sns.countplot(x="Sex", hue="Survived", data=df, palette=palette, hue_order=[0, 1])
graph1.set_yticks(range(0, 1000, 50))
graph1.set_xlabel("Gender", fontsize=12)
graph1.set_ylabel("Number of passengers", fontsize=12)
graph1.set_title("Survival by gender", fontsize=14, fontweight="bold")
plt.show()

# Graph 2  Survival by passenger class
plt.figure()
graph2 = sns.countplot(x="Pclass", hue="Survived", data=df, palette=palette, hue_order=[0, 1])
graph2.set_yticks(range(0, 1000, 50))
graph2.set_xlabel("Passenger class", fontsize=12)
graph2.set_ylabel("Number of passengers", fontsize=12)
graph2.set_title("Survival by passenger class", fontsize=14, fontweight="bold")
plt.show()

# Graph 3  Female survival by passenger class
plt.figure()
graph3 = sns.countplot(x="Pclass", hue="Survived", data=df[df["Sex"] == "female"], palette=palette, hue_order=[0, 1])
graph3.set_yticks(range(0, 700, 50))
graph3.set_xlabel("Passenger class", fontsize=12)
graph3.set_ylabel("Number of women", fontsize=12)
graph3.set_title("Female survival by passenger class", fontsize=14, fontweight="bold")
plt.show()

# Graph 4  Age distribution by survival status
plt.rcParams["font.family"] = "Times New Roman"
palette = {0: "lightskyblue", 1: "lightpink"}

plt.figure()
graph4 = sns.histplot(
    data=df,
    x="Age",
    hue="Survived",
    bins=25,
    palette=palette,
    hue_order=[0, 1],
    alpha=0.7
)

graph4.set_xlabel("Age", fontsize=12)
graph4.set_ylabel("Number of passengers", fontsize=12)
graph4.set_yticks(range(0, 100, 5))
graph4.set_title("Age distribution by survival status", fontsize=14, fontweight="bold")

plt.show()

# Graph 5  Correlation heatmap
cols = ["Survived", "Age", "Fare", "Pclass"]
plt.figure(figsize=(6, 5))
graph5 = sns.heatmap(df[cols].corr(), annot=True, cmap="coolwarm")
graph5.set_title("Correlation heatmap (selected features)", fontsize=14, fontweight="bold")
plt.show()

# Average ticket price for survivors and non-survivors
print("Average ticket price by survival status:")
print(df.groupby("Survived")["Fare"].mean())

# Correlation matrix for selected numerical features
cols = ["Survived", "Age", "Fare", "Pclass"]
print(df[cols].corr())

```
