```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
df = pd.read_csv(r"C:\Users\lesha\Desktop\data science\train.csv")

#Data preparation
avg_age=df["Age"].mean()
df["Age"]=df["Age"].fillna(avg_age)

df["Sex"]=df["Sex"].map({"male":1, "female":0})

from sklearn.model_selection import train_test_split

#feature engineering
df["FamilySize"]=df["SibSp"]+df["Parch"]+1
def family_group(size):
    if size == 1:
        return 0          # "alone"
    elif 2 <= size <= 4:
        return 1          # "small family"
    else:
        return 2          # "big family"

df["FamilyGroup"] = df["FamilySize"].apply(family_group)
X = df[[ "Pclass", "Sex", "Age", "Fare", "FamilyGroup"]]

from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

y=df["Survived"]
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    random_state=42
)



#learning model
machine_learn_model=LogisticRegression(max_iter=1000)
machine_learn_model.fit(X_scaled_train, y_train)

y_pred=machine_learn_model.predict(X_scaled_test)


#estimating of model's accuracy by accuracy score
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)

#estimating model's accuracy by confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import classification_report 
print(classification_report(y_test, y_pred))



```
